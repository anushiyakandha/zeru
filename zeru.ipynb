```python
# Convert columns to numeric
df['value'] = pd.to_numeric(df['value'], errors='coerce')
df['gas_spent'] = pd.to_numeric(df['gas_spent'], errors='coerce')

# Total value moved per wallet
total_value = df.groupby('wallet_id')['value'].sum()

# Number of transactions
tx_count = df.groupby('wallet_id')['tx_hash'].count()

# Total gas spent
total_gas = df.groupby('wallet_id')['gas_spent'].sum()

# Combine into features dataframe
features = pd.DataFrame({
    'total_value': total_value,
    'tx_count': tx_count,
    'total_gas': total_gas
}).fillna(0)

# Normalize and create risk score (0-1000)
features['score'] = (
    (features['total_value'] / features['total_value'].max()) * 400 +
    (features['tx_count'] / features['tx_count'].max()) * 300 +
    (features['total_gas'] / features['total_gas'].max()) * 300
).fillna(0).astype(int)

# Save scores
features[['score']].to_csv('wallet_scores.csv')
print("wallet_scores.csv created")
```

```python
# Convert fees_paid and gas_spent to numeric
df['fees_paid'] = pd.to_numeric(df['fees_paid'], errors='coerce')
df['gas_spent'] = pd.to_numeric(df['gas_spent'], errors='coerce')

# Group by wallet_id
features = df.groupby('wallet_id').agg({
    'fees_paid': 'sum',
    'gas_spent': 'sum',
    'tx_hash': 'count'
}).rename(columns={'tx_hash': 'tx_count'}).fillna(0)

# Add derived feature
features['fees_per_tx'] = features['fees_paid'] / (features['tx_count'] + 1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled = scaler.fit_transform(features)
features_scaled = pd.DataFrame(scaled, index=features.index, columns=features.columns)

# Risk score formula (adjust weights)
features_scaled['risk_score'] = (
    features_scaled['fees_paid'] * 0.5 +
    features_scaled['gas_spent'] * 0.3 +
    features_scaled['tx_count'] * 0.2
) * 1000

final_scores = features_scaled[['risk_score']].reset_index()
final_scores.columns = ['wallet_id', 'score']
final_scores.to_csv("wallet_risk_scores.csv", index=False)

print("wallet_risk_scores.csv created")
```

Brief Explanation
Data Collection Method:
Transaction data for each wallet was retrieved using the Covalent API, focusing on interactions with the Compound V2/V3 lending protocols. This included detailed transaction records such as fees paid, gas spent, and transaction counts.

Feature Selection Rationale:
Key features were selected to reflect wallet activity and risk exposure:

fees_paid: Total transaction fees indicating wallet activity and potential risk.

gas_spent: Gas consumed, reflecting transaction complexity and intensity.

tx_count: Number of transactions representing wallet usage frequency.

Scoring Method:
Features were normalized using Min-Max scaling to standardize their ranges. A weighted sum formula was applied to calculate the risk score:

fees_paid (weight 0.5)

gas_spent (weight 0.3)

tx_count (weight 0.2)
The result was scaled to a 0-1000 range as the final wallet risk score.

Justification of Risk Indicators:
Higher fees and gas spent suggest more complex or frequent interactions with lending protocols, which increases exposure to risk. Transaction count also indicates active wallets that may carry higher risk. These indicators collectively provide a scalable, quantitative measure of wallet risk.
